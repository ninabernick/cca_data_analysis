```python
def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
def plot_subplots_nx2_grid(dataset, variable,n=2):
    cols = ['b', 'orange', 'purple', 'c', 'm', 'y', 'gray', 'pink', 'navy']
    fig, ax = plt.subplots(n, 2, sharex='col', sharey='row', figsize=(7, 5))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    for index, x in enumerate(dataset[variable].values):
        x_index = index %2
        y_index = int(index/n)
        color = cols[index]
        ax[x_index, y_index].loglog(dataset.n_per_ftr1, dataset.sel({'alg':'cca', variable:x}), linestyle='--', c=color)
        ax[x_index, y_index].loglog(dataset.n_per_ftr1, dataset.sel({'alg':'c3a', variable:x}), c=color)
        ax[x_index, y_index].set_ylim(None, 1)
        ax[x_index, y_index].set_title("{}={}".format(variable, x))
    return cols, fig, ax
```

```python
colors, fig, ax = plot_subplots_nx2_grid(ds_px, "px", 2)
lines = [Line2D([0], [0], linewidth=3, linestyle='--', c='b'), Line2D([0], [0], linewidth=3, c='b')]
labels = ["CCA", "C3A"]
fig.legend(lines, labels, bbox_to_anchor=(1.05, .5))
fig.legend(lines, labels, bbox_to_anchor=(1.05, .5))
fig.suptitle('C3A Capability by Dimension of Shared Feature Set')
fig.supxlabel('Samples / feature')
fig.supylabel("$X$ weight error")
```
