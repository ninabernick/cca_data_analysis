Notes on data variables:
- Predicted association strength by the method is stored in `corrs`. The ground truth corr fed into GEMMR is stored in `true_corrs`. 
- `rel_assoc_error` is the relative error of the predicted correlation.
- `cmntrth_weight_error` is the error for X and Y (`modality='x'` or `modality='y'`) predicted weight vectors compared to those predicted by the common truth method (in this case, CCA with known population covariance matrices)
- `cmntrth_loading_error_pearson` could be interesting but currently is all `nan`s

Plot predicted association strength over n_per_ftr2, true association strength, and exa_mix.

```python
ds_r_true = ds.corrs.sel(mix_component=-1, px=16, pya=16, pyb=16, exa_mix=.9, ax=-1, aya=-1, ayb=-1)
fig, ax = plt.subplots(3,2, sharex=True, figsize=(4, 5))


for index, r in enumerate(ds_r_true.rxa.values):
    ax[index, 0].loglog(ds_r_true.n_per_ftr1, ds_r_true.sel(alg='cca', rxa=r, rxb=r, n_per_ftr2=32), c='tab:blue')
    ax[index, 0].plot(ds_r_true.n_per_ftr1, ds_r_true.sel(alg='c3a', rxa=r, rxb=r, n_per_ftr2=32), c='orange')
    ax[index, 0].set_title("r={}".format(r))
    ax[index, 0].set_box_aspect(1/1.5)
    ax[index, 0].hlines(y=r, xmin=0, xmax=256, linestyle='--', linewidth=1, color='purple')
    #ax[index, 0].set_ylabel('r={}'.format(r))

    ax[index, 1].loglog(ds_r_true.n_per_ftr1, ds_r_true.sel(alg='cca', rxa=r, rxb=r, n_per_ftr2=128), c='tab:blue')
    ax[index, 1].plot(ds_r_true.n_per_ftr1, ds_r_true.sel(alg='c3a', rxa=r, rxb=r, n_per_ftr2=128), c='orange')
    ax[index, 1].set_title("r={}".format(r))
    ax[index, 1].set_box_aspect(1/1.5)
    ax[index, 1].hlines(y=r, xmin=0, xmax=256, linestyle='--', linewidth=1,color='purple')

lines = [Line2D([0], [0], linewidth=3, c='tab:blue'), Line2D([0], [0], linewidth=3, c='orange'), Line2D([0],[0], linewidth=2, linestyle="--", c='purple')]
labels = ["CCA", "C3A", "r true"]
fig.legend(lines, labels, bbox_to_anchor=(1.1, .5))
fig.supylabel('Predicted Association Strength')
fig.supxlabel('Samples/feature')
fig.suptitle("\n".join(wrap('C3A Performance by True Neural-Behavioral Association and Supplemental Data Set Size', 40)))
ax[0,0].text(.2, 1.5, "n_per_ftr2 = 32", transform=ax[0,0].transAxes, fontsize='large')
ax[0,1].text(.2, 1.5, "n_per_ftr2 = 128", transform=ax[0,1].transAxes, fontsize='large')
fig.tight_layout()
```
